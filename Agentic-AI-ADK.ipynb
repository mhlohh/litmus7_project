```python
import os
import asyncio
from dotenv import load_dotenv
```

loading the enviorment variables from .env file in the current directory

```python
load_dotenv()

try:
    if os.getenv("GOOGLE_API_KEY"):
        print("✅Gemini API Key loaded from .env file.")
        print(os.getenv('GOOGLE_API_KEY'))
    else:
        print("❌GOOGLE API KEY is not found!")
    
except Exception as e:
    print(f"🔑Setup Error: Details {e}")
```

importing modules essential for building an simple llm agent

```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types
```

HTTP Retry Configuration

```python
retry_config=types.HttpRetryOptions(
    attempts=3,
    exp_base=2,
    initial_delay=1,
    http_status_codes=[429,500,503,504]
)
```

Intializing the agent and setup

```python
root_agent = Agent(
    model='gemini-2.0-flash-lite',
    name='root_agent',
    description='A helpful assistant for user questions.',
    instruction='Answer user questions to the best of your knowledge',
    tools=[google_search]
)
print("✅Agent is Created")
```

main asynchronous function 

```python
async def main():
    runner = InMemoryRunner(agent=root_agent)
    print("Chat with the agent (type 'q' to quit):")
    while True:
        prompt = input("$: ")
        if prompt.strip().lower() == 'q':
            print("Goodbye!")
            break
        if not prompt.strip():
            continue
        await runner.run_debug(prompt)
```

calling main function

```python
try:
    import nest_asyncio
    nest_asyncio.apply()
except Exception:
    pass
await main()
import asyncio
loop = asyncio.get_event_loop()
task = loop.create_task(main())
await task
```
